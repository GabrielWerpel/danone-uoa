# Data Exploration

```python
import os
import pandas as pd
```

Load data and quick preview

```python
csv_file_path = "DATA_101549999 Can Weight Report.csv"
df = pd.read_csv(csv_file_path)
df.head()
```

```python
columns_to_check = ['USER', 'PRODUCTION ORDER NUMBER', 'SKU', 'FORMULATION', 'AVERAGE TARE WEIGHT', 'CAN NET WEIGHT', 'DENOMINATION']
unique_counts = df[columns_to_check].nunique()
print(unique_counts)
```

```python
df_weight = df[['DATE', 'TIME', 'PRODUCTION ORDER NUMBER', 'AVERAGE TARE WEIGHT', 'CAN NET WEIGHT', 'DENOMINATION']]
df_weight.head()
```

```python
df_weight['DateTime'] = pd.to_datetime(
        df_weight['DATE'] + ' ' + df_weight['TIME']
    ).dt.tz_localize(
        'Pacific/Auckland'
    ).dt.tz_convert('UTC')

df_weight.head()
```
